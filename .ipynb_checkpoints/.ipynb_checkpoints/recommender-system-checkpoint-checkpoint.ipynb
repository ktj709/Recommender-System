##### Content based filtering - In this algo, we try finding item look alike. Once we have item look alike matrix, we can easily recommend alike items to a customer who has purchased any item from the source.

#####  Collaborative filtering - Here, we try to search for lookalike has customers and offer products based on what his/her lookalike has chosen. This algorithm is very effective but takes a lot of time and resources.

##### Hybrid filtering - Using both content based and collaborative filtering. e.g. - FACEBOOK

```python
ratings=pd.read_csv('Ratings.csv')
```

```python
ratings.head(2)
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'rating'},inplace=True)
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books=ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.shape
```

```python
number_rating = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'number of ratings'},inplace=True)
```

```python
number_rating
```

```python
final_rating = ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['number of ratings']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
# to extract a row, use iloc function
distances, suggestions = model.kneighbors(book_pivot.iloc[237].values.reshape(1, -1), n_neighbors=6)
```

```python
book_pivot
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
book_pivot.index[237]
```

```python
np.where(book_pivot.index=='Animal Farm')[0][0]
```

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)

    for i in range(len(suggestions)):
        if i==0:
            print("The suggestions for", book_name,"are :")

        if not i:
            print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Animal Farm')
```
